## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

```python
def check_age_below_min(applicant_age):
    """
    Checks if the applicant's age is below the minimum required age of 18.

    Args:
        applicant_age (int): The age of the applicant.

    Returns:
        dict or None: A dictionary with 'code' and 'description' if the rule is triggered,
                      otherwise None.
    """
    if applicant_age < 18:
        return {
            'code': 'AGE_BELOW_MIN',
            'description': "Applicant's age is below the minimum required age of 18."
        }
    return None

# --- Demonstration of the function ---
print("Checking applicants:")

applicant1_age = 17
reason1 = check_age_below_min(applicant1_age)
if reason1:
    print(f"Applicant 1 (Age: {applicant1_age}): Rule triggered - {reason1['code']}: {reason1['description']}")
else:
    print(f"Applicant 1 (Age: {applicant1_age}): No rule triggered.")

applicant2_age = 20
reason2 = check_age_below_min(applicant2_age)
if reason2:
    print(f"Applicant 2 (Age: {applicant2_age}): Rule triggered - {reason2['code']}: {reason2['description']}")
else:
    print(f"Applicant 2 (Age: {applicant2_age}): No rule triggered.")

applicant3_age = 18
reason3 = check_age_below_min(applicant3_age)
if reason3:
    print(f"Applicant 3 (Age: {applicant3_age}): Rule triggered - {reason3['code']}: {reason3['description']}")
else:
    print(f"Applicant 3 (Age: {applicant3_age}): No rule triggered.")
```
